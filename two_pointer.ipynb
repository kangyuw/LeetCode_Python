## [42. Trapping Rain Water](https://leetcode.com/problems/trapping-rain-water/)

### 2 Pointers

* left: start from left to right, finding maxL

* right: start from right to left, finding maxR

* if left < right: moving left foward

* 2 higher bar can trap rain water

* the maximum of water height: the bar with lower height within 2 higher bars

* trap rain = the lower bar - even lower bar

...

```python
class Solution:
    def trap(self, height: List[int]) -> int:
        if not height:
            return 0
        left = 0
        right = len(height) - 1
        res = 0
        maxL = height[left]
        maxR = height[right]

        while left < right:
            if maxL < maxR:
                left += 1
                maxL = max(maxL, height[left])
                res += maxL - height[left]
            else:
                right -= 1
                maxR = max(maxR, height[right])
                res += maxR - height[right]
        return res
```

## [11. Container With Most Water](https://leetcode.com/problems/container-with-most-water/)

### 2 Pointers

如果 left < right: 將 left += 1, 尋找可能更大的結果

否則: 將 right -= 1, 尋找可能更大的結果

...

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        n = len(height)
        left = 0
        right = n - 1
        res = 0
        while left < right:
            res = max(res, min(height[left], height[right]) * (right - left))
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return res
```

## [3Sum](https://leetcode.com/problems/3sum/)

* if the combination is too small: moving "left" to the right

* if the combination is too large: moving "right" to the left

遍歷矩陣, 對後面的項目進行 sorted two sum (2 pointers)

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        # find three numbers combine to 0
        # first acquire one number then operate two sum
        res = []
        nums.sort()
        n = len(nums)
        for i in range(n-2):
            # afterward its all positive
            if nums[i] > 0:
                break
            if i > 0 and nums[i] == nums[i-1]:
                continue
            left, right = i+1, n-1
            # sorted twoSum
            while left < right:
                comb = nums[i] + nums[left] + nums[right]
                if comb < 0:
                    left += 1
                elif comb > 0:
                    right -= 1
                else:
                    res.append([nums[i], nums[left], nums[right]])
                    # prevent repeat
                    while left < right and nums[left] == nums[left+1]:
                        left += 1
                    while left < right and nums[right] == nums[right-1]:
                        right -= 1
                    left += 1
                    right -= 1
        
        return res
```

## [763\. Partition Labels](https://leetcode.com/problems/partition-labels/)

每個字母只能出現在一個子字串裡面

盡量分割成越多子字串越好

### Brute Force

遍歷矩陣, 從第一次出現此字母, 到最後一次出現此字母, 建立一個範圍

如果超過此範圍就擴充它

如果遍歷到範圍底, 切割

Time: $O(N^2)$

### 2 pointers with Hash Table

* start: start of the next partition

* end: end of current partition, which is decide when the current index is the last appearance index

使用 hast table 儲存每個字母出現的最後一個位置

然後再遍歷矩陣, 如果最後出現的位置大於目前的範圍, 擴充

如果遍歷到目前範圍的最後一個座標, 切割

```python
class Solution:
    def partitionLabels(self, S: str) -> List[int]:
        ## store the last appear index of each charcters
        last = {c: i for i, c in enumerate(S)}
        
        start, end = 0, 0
        res = []
        for i, c in enumerate(S):
            end = max(end, last[c])
            if i == end:
                res.append(end - start + 1)
                start = end+1
        return res
```

## [234. Palindrome Linked List](https://leetcode.com/problems/palindrome-linked-list/)

### 2 Pointers

* 當 fast 走到底時, slow 會在中心點

* 使用 tail, 從 slow 開始將後半段倒反

* 比較 tail 及 head

...

```python
class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        fast = slow = head
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next
        
        ## reverse later-half of the linked list
        tail = None
        while slow:
            nxt = slow.next
            slow.next = tail
            tail = slow
            slow = nxt
        ## compare tail and head
        while tail:
            if tail.val != head.val:
                return False
            tail = tail.next
            head = head.next
        
        return True
```

## [88. Merge Sorted Array](https://leetcode.com/problems/merge-sorted-array/)

### 2 pointers

pointer 從陣列的後面開始往前記數

p 標示在 nums1 中即將被取代的位置

較大者填進 p 的位置, 並且往前遍歷

最後將 nums2 剩餘的陣列合併在 nums1 的前面

...

```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        ptr1 = m-1
        ptr2 = n-1
        p = m+n-1
        while ptr1 >= 0 and ptr2 >= 0:
            if nums1[ptr1] < nums2[ptr2]:
                nums1[p] = nums2[ptr2]
                ptr2 -= 1
            else:
                nums1[p] = nums1[ptr1]
                ptr1 -= 1
            p -= 1
        
        nums1[:ptr2+1] = nums2[:ptr2+1]
```

## [532. K-diff Pairs in an Array](https://leetcode.com/problems/k-diff-pairs-in-an-array/)

回傳相差為 k 的 pair 總共有幾個

### 2 pointers

整理陣列

* 當 left 與 right 相等, 或是差值較小: right += 1

* 當差值較大: left += 1

* 當差值相等: res += 1, left 前進到下一個數字

time: $O(NlogN)$

space: $O(N)$

```python
class Solution:
    def findPairs(self, nums: List[int], k: int) -> int:

        nums = sorted(nums)

        left = 0
        right = 1
        res = 0

        while (left < len(nums) and right < len(nums)):
            if (left == right or nums[right] - nums[left] < k):
                right += 1
            elif nums[right] - nums[left] > k:
                left += 1
            else:
                res += 1
                left += 1
                while (left < len(nums) and nums[left] == nums[left - 1]):
                    left += 1

        return res
```
