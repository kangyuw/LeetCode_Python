## [94. Inorder Traversal](https://leetcode.com/problems/binary-tree-inorder-traversal/submissions/)

left &rarr; root &rarr; right

stack + 迴圈

往左走, 把經過的點丟進 stack 中

走到底後, 開始退 stack, 往右走

```c++=
class Solution {
public:
    vector<int> inorderTraversal(TreeNode* root) {
        vector<int> inorder;
        stack<TreeNode*> nodes;
        
        if(root == NULL) return inorder;
        
        while(root != NULL || !nodes.empty()){
            //push left children if available
            while(root != NULL){
                nodes.push(root);
                root = root->left;
            }
            
            //retrieve top node and store its right child if exists
            root = nodes.top();
            nodes.pop();
            
            inorder.push_back(root->val);
            root = root->right;
        }
        
        return inorder;
    }
};
```

````
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        inorder, stack = [], []
        while root or stack:
            while root:
                stack.append(root)
                root = root.left
            node = stack.pop()
            inorder.append(node.val)
            root = node.right
        return inorder
````

## [98. Validate](https://leetcode.com/problems/validate-binary-search-tree/) 52%

對 BST 來說, inorder 遍歷後就是從小到大排序

在 Python 中 0 會被當作不存在, 必須要判定 None

```c++=
class Solution {
public:
    bool isValidBST(TreeNode* root) {
        if(root == NULL) return true;
        
        stack<TreeNode*> s;
        TreeNode* prev = NULL;
        
        while(root != NULL || !s.empty()){
            while(root != NULL){
                s.push(root);
                root = root->left;
            }    
            
            root = s.top(); s.pop();
            if(prev != NULL && prev->val >= root->val) return false;
            prev = root;
            root = root->right;
        }
        
        return true;
    }
};
```

````
class Solution():
    def isValidBST(self, root):
        if not root: return True
    
        stack = []
        prev = TreeNode(None)

        while root or stack:
            while root:
                stack.append(root)
                root = root.left
            
            root = stack.pop()
            if prev.val != None and prev.val >= root.val: return False;
            prev = root
            root = root.right

    
        return True
````

### 思路2: 上下限 95%

對 left 而言, root 應該要比較大

對 right 而言, root 應該要比較小

````
class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        return self.helper(root, float('-inf'), float('inf'))
    
    def helper(self, root, lower, higher):
        # if we got to a leaf then we know we got to the end and its true 
        if not root:
            return True
        # along the way, if lower exists, and  return false 
        # we check 
        if lower is not None and root.val <= lower:
            return False 
        
        # along the way if we hit bad return false 
        if higher is not None and root.val >= higher:
            return False 
        
        # if we havent hit anything keep looking 
        return self.helper(root.left, lower, root.val) and self.helper(root.right, root.val, higher)
````

## [144. Binary Tree Preorder Traversal](https://leetcode.com/problems/binary-tree-preorder-traversal/)

```c++

class Solution {
public:
    // DFS magic : initialize stack and do the following 
    // pop top
    // retrieve neighbours for top
    // push unvisited neighbours to stack | repeat while stack not empty 
    // because this is a tree no need to keep track of visited as no cycles possible. 
    
    vector<int> preorderTraversal(TreeNode* root) {
        stack<TreeNode*> s;
        vector<int> result;
        
        s.push(root);
        
        while(!s.empty()){
            TreeNode* current = s.top(); s.pop();  // pop top 
            
            if(current != NULL){
                // push unvisited neighbours to stack | order matters here, if you reverse it 
                // it would still be a DFS but a symmetric one to preorder out of the 6 possible combinations.  
                s.push(current->right);
                s.push(current->left);
                
                result.push_back(current->val);
            }
        }
        
        return result;
    }
};
```

````
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        res, stack = [], [root]
        while stack:
            cursor = stack.pop()
            if cursor:
                stack.append(cursor.right)
                stack.append(cursor.left)
                res.append(cursor.val)
        return res
````

## [323. Number of Connected Components in an Undirected Graph](https://www.jianshu.com/p/287e61b9237c)

輸入
+ n: 點的個數
+ edges: 一對點

輸出
+ 有幾組無向圖?

針對每個節點做 DFS, 把沒有造訪過的放進 stack 中

stack 空了, DFS 結束, 答案 + 1

```c++
class Solution {
public:
    int countComponents(int n, vector<pair<int, int>>& edges) {
        vector<bool> visited(n, false);
        vector<vector<int>> adjList(n, vector<int>(0));
        stack<int> dfs;
        
        int count = 0;
        int ans = 0;
        
        // building the graph's adjacency list
        for(auto edge : edges){
            int from = edge.first;
            int to   = edge.second;
            
            adjList[from].push_back(to);
            adjList[to].push_back(from);
        }
        
        // dfs on nodes to find connected components
        for(int i = 0; i < n; i++){  
            if(!visited[i]){
                ans++;
                dfs.push(i);
                
                while(!dfs.empty()){
                    int current = dfs.top(); dfs.pop();
                    visited[current] = true;
                    
                    for(int neighbour : adjList[current]){
                        if(!visited[neighbour]) dfs.push(neighbour);
                    }
                }
            }             
        }
        
        return ans;
    }
};
```

## [200. Number of Islands](https://leetcode.com/problems/number-of-islands/)

1 是陸地 0 是海 有幾座島?

```c++

class Solution {
public:
    int numIslands(vector<vector<char>>& grid) {
        int ans = 0; // number of groups 
      
        // iterating through given grid to find a '1'
        for(int i = 0; i < grid.size(); i++){
            for(int j = 0; j < grid[0].size(); j++){
                if(grid[i][j] == '1'){
                    ans++; // start a group and visit all members of this group using dfs
                    dfs(grid, i, j);
                }
            }
        }
        
        return ans;
    }
    
    void dfs(vector<vector<char>>& grid, int y, int x){
        // if out of bounds or at a cell with '0' or '*', simply stop and return | end the dfs
        if(x < 0 || x >= grid[0].size() || y < 0 || y >= grid.size() || grid[y][x] != '1') return;
        
        grid[y][x] = '*'; // destroying visited cells so they are not re-visited
        
        // recursive dfs on all neighbours
        dfs(grid, y + 1, x);
        dfs(grid, y - 1, x);
        dfs(grid, y, x + 1);
        dfs(grid, y, x - 1);
    }
};
```


### 思路1: 遞迴DFS

每個點有四個方向, 造訪過變成 0

碰到 1 把座標放進 stack 中

````
class Solution:
    def dfs(self, grid, y, x):
        if x < 0 or x >= len(grid[0]) or 
           y < 0 or y >= len(grid) or grid[y][x] != '1':
            return

        grid[y][x] = '*'
        self.dfs(grid, y+1, x)
        self.dfs(grid, y-1, x)
        self.dfs(grid, y, x+1)
        self.dfs(grid, y, x-1)
    
    def numIslands(self, grid: List[List[str]]) -> int:
        res = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == '1':
                    res += 1
                    self.dfs(grid, i, j)
        return res
````

### 思路2: DFS迴圈

````
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        res = 0
        if grid:
            for r in range(len(grid)):
                for c in range(len(grid[0])):
                    if grid[r][c] == '1':
                        stack = [[r, c]]
                        while stack:
                            [x, y] = stack.pop()
                            grid[x][y] = '0' # visist後設為 0
                            if x+1 < len(grid) and grid[x+1][y] == '1':
                                stack.append([x+1, y])
                            if x-1 >= 0 and grid[x-1][y] == '1':
                                stack.append([x-1, y])
                            if y+1 < len(grid[0]) and grid[x][y+1] == '1':
                                stack.append([x, y+1])
                            if y-1 >= 0 and grid[x][y-1] == '1':
                                stack.append([x, y-1])
                        res += 1
        return res
````

## [547. Friend Circles](https://leetcode.com/problems/friend-circles/)

輸入:
+ 矩陣[ i ][ j ] == 1: i 跟 j 之間互為朋友

只要有互為朋友的人, 間接直接, 都算在同一個朋友圈中

計算總共有多少朋友圈

### 思路1: 迴圈 DFS 30%
```c++=
class Solution {
public:
    // M is the friend matrix provided | basic intuition: we need connected components in graph, which can be done by BFS / DFS.
    // why does this work ? because applying transitive property to this : a friend of b, b friend of c , c friend of d
    // a,b,c,d should fall into 1 friend circle, just imagine a,b,c,d to be nodes of graph and they are connected by an edge if they 
    // are friends else not connected by an edge, so simply finding the total groups / connected components gives us the answer. 
    int findCircleNum(vector<vector<int>>& M) {                                     
        vector<bool> friendZoned(M.size(), false);                                  // array to keep track of visited in DFS
        stack<int> dfs;                                                             // stack to execute DFS 
        
        int circles = 0;                                                            // final ans stored in circles
        
        for(int i = 0; i < M.size(); i++){
            if(!friendZoned[i]){                                                   
                circles++;  // // person i unvisited so start new friend circle
                
                // DFS magic : { push to stack - pop top - retrieve neighbours - repeat first 3 steps for each unvisited neighbour until stack empty } 
                dfs.push(i);
                
                while(!dfs.empty()){
                    int current = dfs.top(); dfs.pop(); // pop top 
                    friendZoned[current] = true;        // mark node visited
                    
                    // retrieve it's unvisited neighbours and push them to stack
                    for(int j = 0; j < M[current].size(); j++){
                        if(!friendZoned[j] && M[current][j] == 1){                                
                            dfs.push(j);
                        }
                    }
                }
            }
        }
        
        return circles;
    }
};
```

把 0 號所有未造訪的朋友放進 stack 中

把 0 號標為已造訪

stack.pop()

做完一次 DFS, res += 1

找下一個未造訪的同學做 DFS

````
class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        visited = [False]*len(M)
        stack = []
        res = 0

        for i in range(len(M)):
            if not visited[i]:
                res += 1

                stack.append(i)
                while stack:
                    cursor = stack.pop()
                    visited[cursor] = True
                    for j in range(len(M[cursor])):
                        if not visited[j] and M[cursor][j]:
                            stack.append(j)
        return res
````

### 思路2: DFS迴圈 99%

差別1: visited 使用 set 而非 list

差別2: 每次輸入一列, 使用 enumerate, 少一次迴圈

````
class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        visited = set()
        def dfs(row):
            for stu, fri in enumerate(row): // 回傳 (同學的編號, 是不是朋友)
                if fri and stu not in visited: // 是朋友, 且沒有造訪過
                    visited.add(stu)
                    dfs(M[stu])
        res = 0
        for i in range(len(M)):
            if i not in visited:
                dfs(M[i])
                res += 1
        return res
````

## [102. Binary Tree Level Order Traversal](https://leetcode.com/problems/binary-tree-level-order-traversal/)

回傳得要分層的

### 思路1: 兩個迴圈 70%

第一個迴圈: BFS 的 Queue

第二個迴圈: 每一層的長度


````
from collections import deque
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        if not root:
            return []
        queue, res = deque([root]), []
        while queue:
            cur_level, size = [], len(queue)
            for i in range(size):
                node = queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                cur_level.append(node.val)
            res.append(cur_level)
        return res

class Solution_Alt:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        result = []
        
        if root == None:
            return result
        queue = [root]
        while len(queue) > 0:
            items_in_row = len(queue)
            row_values = []
            while items_in_row > 0:
                node = queue.pop(0)
                row_values.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                items_in_row = items_in_row - 1
            result.append(row_values)
        return result
````

### 思路2: dummy node 70%

用 dummy 標示一層的結束

````
from collections import deque
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        res = []
        level_res = []
        bfs = deque()
        marker = TreeNode(float("-inf"))

        bfs.append(root)
        bfs.append(marker)

        while(bfs):
            current = bfs.popleft()

            if(current):
                if(current.val == float("-inf")): # 碰到 marker 了
                    res.append(level_res)
                    level_res = []
                    if(bfs): bfs.append(marker)
                else:
                    level_res.append(current.val)
                    bfs.append(current.left)
                    bfs.append(current.right)
                    
        res.pop()
        return res
````

### 思路3: 兩個 Queue 70%

奇數 queue + 偶數 queue

奇數在左, 偶數在右

root 是第0層, 放進偶數 queue 中

偶數 queue pop 時, 將 pop 出的左子及右子放進奇數 queue 中

````
from collections import deque
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        res = []
        even = deque()
        odd = deque()
        
        even.append(root)

        while even or odd:
            level = []

            if not odd:
                while even:
                    curr = even.popleft()
                    if curr:
                        level.append(curr.val)
                        odd.append(curr.left)
                        odd.append(curr.right)

            elif not even:
                while odd:
                    curr = odd.popleft()
                    if curr:
                        level.append(curr.val)
                        even.append(curr.left)
                        even.append(curr.right)
            
            res.append(level)
        res.pop()
        return res
````

````
class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
        if root is None:
            return True
        return self.areSymmetric(root.left, root.right)
        
    def areSymmetric(self, left: TreeNode, right: TreeNode):
        if left is None:
            return right is None
        if right is None:
            return False
        if left.val != right.val:
            return False
        return self.areSymmetric(left.left, right.right) and self.areSymmetric(left.right, right.left)
````

## [542. 01 Matrix](https://leetcode.com/problems/01-matrix/)

回傳每個位置與最近的0的距離

### 思路1: BFS top-down 25%

從0開始, 遍歷矩陣中所有0的元素 &rarr; root

以下開始 bfs

然後把旁邊尚未造訪過的點 (1的點), 標記為 1 &rarr; level 1

繼續, 把尚未造訪過的點標記為 2 &rarr; level 2

以此類推

### 結論

1. 最短路徑題目: 試試 BFS

2. 如果是邊界一樣的圖, 找最短路徑: 試試 BFS, 不需要 Dijkstra

3. 走迷宮就是比較炫泡的 graph 而已: 試試 BFS

````
class Solution:
    def updateMatrix(self, matrix: List[List[int]]) -> List[List[int]]:
        bfs = []
        m = len(matrix)
        n = len(matrix[0])

        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    bfs.append([i,j]) # 遍歷所有的 0
                else:
                    matrix[i][j] = float('inf') # 不是0就設定成無限
        direction = [[1,0],[-1,0],[0,1],[0,-1]]

        while bfs:
            current = bfs.pop(0)

            for d in direction:
                i = current[0] + d[0]
                j = current[1] + d[1]

                if i<0 or i>=m or j<0 or j>=n or matrix[i][j] <= matrix[current[0]][current[1]] + 1: 
                    continue # 撞到邊界 or 沒有比較近
                bfs.append([i,j])
                matrix[i][j] = matrix[current[0]][current[1]] + 1
        return matrix
````

### 思路2: BFS bottom-up 99%

1. 善用 enumerate 可以簡短運算時間

2. 多維矩陣就用多層的 enumerate 來記錄座標

3. 可以不用 list 就不要用

遍歷所有的 1

````
class Solution:
    def updateMatrix(self, matrix: List[List[int]]) -> List[List[int]]:
        if not any(matrix): # 如果全為 0
            return matrix
        
        len_row = len(matrix)
        len_col = len(matrix[0])
        queue = collections.deque()
        level = 1
        
        for r, row in enumerate(matrix): # 遍歷所有的 1
            for c, val in enumerate(row):
                if val:
                    queue.append((r, c))
                    
        while queue:
            count = len(queue)
            while count: # 只需要做
                r, c = queue.popleft() # 1 的座標
                if (r == 0 or matrix[r-1][c] >= level) \ # 已經是最上 or 上面 >= level
                and (c == 0 or matrix[r][c - 1] >= level) \ # 已經是最左 or 左邊 >= level
                and (len_row == r + 1 or matrix[r + 1][c] >= level) \ # 已經是最下 or 下面 >= level
                and (len_col == c + 1 or matrix[r][c + 1] >= level): # 已經是右 or 右邊 >= level
                    matrix[r][c] = level + 1 # 標示為下一個 level 要處理的
                    queue.append((r, c)) # 放進 queue 看下次能不能過關
                count -= 1
            level += 1 # 目前檢視的 level
            
        return matrix
````

````
import collections, sys
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # 初始化字母表
        table = collections.Counter(t)

        # 初始化 Sliding Window
        begin = 0
        end = 0
        counter = len(table)
        cur_len = sys.maxsize
        ans=""

        # Start Sliding Window
        while end < len(s):
            endchar = s[end]

            # 在字母表中找到對應的字母
            if endchar in table:
                table[endchar] -= 1
                if table[endchar] == 0:
                    counter -= 1
            end += 1

            # 如果新 Window 比之前都要小
            while counter == 0:
                if end - begin < cur_len:
                    cur_len = end - begin
                    ans = s[begin:end]

                # 如果 s 開頭的字元 不在字母表中: 去掉它
                startchar = s[begin]

                if startchar in table:
                    table[startchar] += 1
                    if table[startchar] > 0:
                        counter += 1

                begin += 1
            print(begin, end, table)

        return s[begin:end]
````

## 更快的做法 99%

不需要匯入 sys, collections

counter = len(t)

````
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        n, counter= len(s), len(t)
        start = end = 0
        dic = {}
        lenMin = n+1
        head = 0
        
        for c in t:
            if c in dic:
                dic[c] += 1
            else:
                dic[c] = 1

        while end < n:
            if s[end] in dic:
                if dic[s[end]] > 0:
                    counter -= 1
                dic[s[end]] -= 1
                
            end += 1
            
            while counter == 0:
                if end - start < lenMin:
                    lenMin = end-start
                    head = start               
            
                if s[start] in dic:
                    dic[s[start]] += 1
                    if dic[s[start]] > 0:
                        counter += 1
                start += 1
        if lenMin == n+1:
            return ""
        return s[head: head+lenMin]
 
````

## [438. Find All Anagrams in a String](https://leetcode.com/problems/find-all-anagrams-in-a-string/)

找出字串 s 中, 所有字串 p 可以重新排列組合出來的結果

輸出字串 s 的座標

長度固定的 window

### 思路1:
每次都從 s 中切出 長度固定的段落跟 p 比較

Time: $O(N^2)$ 重複計算

### 思路2: 長度固定的 Window 90%

每次都要看開頭的退出的字母 是不是有在 p 之中

Time: $O(N)$

````
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        if len(s) < len(p): return []
        ans = []
        dic = {}
        for c in p:
            if c in dic: dic[c] += 1
            else: dic[c] = 1
        start = end = 0
        n = len(s)
        counter = len(p)
            
        while end < n:
            if s[end] in dic:
                if dic[s[end]] > 0: counter -= 1
                dic[s[end]] -= 1
                
            end += 1
            if counter == 0: ans.append(start)
            if end - start >= len(p):
                if s[start] in dic:
                    dic[s[start]] += 1
                    if dic[s[start]] > 0:
                        counter += 1
                start += 1
        
        return ans
````

## [567. Permutation in String](https://leetcode.com/problems/permutation-in-string/)

### 思路1: 微調 77%

````
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        n = len(s1)
        m = len(s2)
        if n > m: return False
        dic = {}
        for c in s1:
            if c in dic: dic[c] += 1
            else: dic[c] = 1
        start = end = 0
        
        while end < m:
            if s2[end] in dic:
                if dic[s2[end]] > 0: n -= 1
                dic[s2[end]] -= 1
            
            end += 1
            if n == 0: return True
            if end - start >= len(s1):
                if s2[start] in dic:
                    dic[s2[start]] += 1
                    if dic[s2[start]] > 0: n += 1

                start += 1
            
        return False
````

### 思路2: 兩個 dic 互相比較

直接建立兩個 Count Table 互相比較

直接把 window 套在字串上 減少遍歷長度

````
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        
        if len(s2) < len(s1):
            return False
        
        s1_cntr = {chr(i+97): 0 for i in range(26)}
        s2_cntr = {chr(i+97): 0 for i in range(26)}
        
        for char in s1:
            s1_cntr[char] += 1
            
        for char in s2[:len(s1)]:
            s2_cntr[char] += 1
        
        if s1_cntr == s2_cntr:
            return True
        
        for idx in range(len(s2) - len(s1)):
            
            new_idx = idx + len(s1)
            s2_cntr[s2[idx]] -= 1
            s2_cntr[s2[new_idx]] += 1
            
            if s1_cntr == s2_cntr:
                return True
        
        return False
````
