# 1. First Bad Version

壞版本的後面就都是壞版本

給一個任意數字, 找到最找的壞版本

## Solution 1: Binary Search

89.42%

mid 是壞版本: 目標可能更小

mid 是好版本: 目標可能更大

直到 low 就是 mid

```
'''
The isBadVersion API is already defined for you.
@param version, an integer
@return a bool
def isBadVersion(version):
'''
class Solution:
    def firstBadVersion(self, n):
        """
        :type n: int
        :rtype: int
        """
        def search(low,high):
            
            mid = (low+high) // 2
            
            if isBadVersion(mid):
                if mid == low:
                    return mid
                return search(low, mid)
            else:
                return search(mid+1,high)
                
        res = search(1,n)
        return res
```

## Solution 2: Binaray Search


```
class Solution:
    def firstBadVersion(self, n):
        """
        :type n: int
        :rtype: int
        """
        low , high = 0, n
        
        while l <= h:
            
            m = (l+h) // 2
            
            if l == m:            
                if not isBadVersion(m):
                    return m + 1
             
            if isBadVersion(m):
                h = m
            else:
                l = m
        
        return l+1
```
