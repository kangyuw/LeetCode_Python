```
class Solution:
    def floodFill(self, image, sr, sc, newColor):
        queue = [[sr, sc]]
        oc = image[sr][sc]
        image[sr][sc] = newColor
        if oc == newColor:
            return image
        directions = [[1,0],[-1,0],[0,1],[0,-1]]
        while queue:
            current = queue.pop(0)
            for d in directions:
                i = current[0] + d[0]
                j = current[1] + d[1]
                if i < 0 or j < 0 or j >= len(image[0]) or i >= len(image):
                    continue
                if image[i][j] == oc:
                    queue.append([i,j])
                    image[i][j] = newColor
        
        return image
```
