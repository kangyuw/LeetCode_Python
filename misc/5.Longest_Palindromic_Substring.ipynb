### 成功

10.67%

為了避免沒有走到結果 => reversed

dp[ i \][ j \]: 從 i 開始到 j 的字串是回文嗎?

i 從尾部往前遍歷, j 從 i 走到底


````
class Solution:
    def longestPalindrome(self, s: str) -> str:
        n = len(s)
        if n == 1: return s
        elif n == 0: return s
        dp = [ [None for _ in range(n)] for _ in range(n)]
        
        # dp[i][j]: s[i:j+1] is Palin
        
        res = ''
        
        for i in reversed(range(n)):
            for j in range(i,n):
                
                if j - i < 3:
                    dp[i][j] = s[i] == s[j]
                
                else: 
                    dp[i][j] = s[i] == s[j] and dp[i+1][j-1]
            
                if dp[i][j] and j-i+1 >= len(res):
                    res = s[i:j+1]
                    
        return res
````

## expendAroundCenter

75%

i 是每次擴張的中心

如果本身是回文 就往兩側擴張 再繼續看是不是回文

````
class Solution:
    def longestPalindrome(self, s: str) -> str:
        start = end = 0
        n = len(s)
        def expandAroundCenter(left, right, s):
            while left >= 0 and right < n and s[left] == s[right]:
                left-=1
                right+=1
            return right-left-1 # length
        
        for i in range(n):
            oddLen = expandAroundCenter(i, i, s)
            evenLen = expandAroundCenter(i, i+1, s)
            length = max(oddLen, evenLen)
            
            if length > end - start:
                start = i - int((length - 1) / 2)
                end = i + int((length) / 2)
                
        return s[start:end+1]
````

## tail

96%

尋找到第 i 個字元之前, 最長的回文

至少有 maxlen 的長度 再去檢查

slice 超過範圍沒關係

````
class Solution:
    def longestPalindrome(self, s):
        if len(s) < 2 or s == s[::-1]:
            return s
        
        start, maxlength = 0, 1
        for i in range(len(s)):
            odd = s[i-maxlength-1:i+1]
            even = s[i-maxlength:i+1]    
            if i-maxlength-1 >= 0 and odd == odd[::-1]: # 至少有 maxlen 的長度 再去檢查
                start = i-maxlength-1
                maxlength += 2
            elif i-maxlength >= 0 and even == even[::-1]: # 至少有 maxlen 的長度 再去檢查
                start = i-maxlength
                maxlength += 1
        return s[start:start+maxlength]
````
