## Solution 1: DP

DP: M * N

目標
```
DP[i][j] = 以 matrix[i][j] 為最右下角的格子, 能畫出最大正方形的邊長
```
3 的旁邊都會是 2, 2 的旁邊都會是 1

邊長 = 從 1, 2, ... n 的正方形個數

Update:
```
if matrix[i][j] == 1:
    DP[i][j] = min( DP[i-1][j], DP[i][j-1], DP[i-1][j-1]) + 1
```

return: sum(DP)

$O(mn)$

### Improve
因為 dp 只會向前更新, 可以直接 update matrix 本身

````
class Solution:
    def countSquares(self, matrix: List[List[int]]) -> int:
        m = len(matrix[0])
        n = len(matrix)
        res = 0
        dp = [[0 for _ in range(m)] for _ in range(n)]
        for i in range(n):
            for j in range(m):
                dp[i][j] = matrix[i][j]
                if dp[i][j] and i and j:
                    dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) + 1
                res += dp[i][j]
        
        return res
````

````
class Solution:
    def countSquares(self, matrix: List[List[int]]) -> int:
        if not matrix or not matrix[0]: return 0
        n = len(matrix)
        m = len(matrix[0])
        
        for i in range(1,n):  # skip row 0
            for j in range(1,m):  # skip col 0
                if matrix[i][j]:  # do not require extra space
                    matrix[i][j] = 1+min(matrix[i-1][j], matrix[i][j-1], matrix[i-1][j-1])
        return sum([sum(row) for row in matrix])
````
