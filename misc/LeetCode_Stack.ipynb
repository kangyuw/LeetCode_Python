````
class MinStack:

    def __init__(self):
        self.q = []

    # param x, an integer
    # return an integer
    # 總是把最小值存在最外側的其中一格中
    def push(self, x):
        curMin = self.getMin()
        if curMin == None or x < curMin:
            curMin = x
        self.q.append((x, curMin));

    # return nothing
    def pop(self):
        self.q.pop()


    # return an integer
    def top(self):
        if len(self.q) == 0:
            return None
        else:
            return self.q[len(self.q) - 1][0]


    # return an integer
    # 不需要額外的搜尋時間
    def getMin(self):
        if len(self.q) == 0:
            return None
        else:
            return self.q[len(self.q) - 1][1]
````

````
## Fastest Solution 28 ms

def isValid(self, s: str) -> bool:
    stack = []

    symbols = {")": "(", "}": "{", "]": "["}

    for c in s:

        if c in symbols:
            top = stack.pop() if stack else '#'

            if symbols[c] != top:
                return False
        else:
            stack.append(c)


    return not stack
````

# [Daily Temperatures](https://leetcode.com/explore/learn/card/queue-stack/230/usage-stack/1363/)

給個記錄每天天氣的List T 

return 另一個 List 紀錄要等幾天才能等到更暖的天



````
## 320 ms, 17.2 MB
def dailyTemperatures(T):
    if not T: return
    res = [0] * len(T)
    stk = []
    for i in range(len(T)):
        cur = T[i]
        if stk:
            last, idx = stk[-1]
            while stk and cur > last:
                stk.pop()
                res[idx] = i - idx
                if stk: last, idx = stk[-1]
                else: last = None
        stk.append((cur, i))
    return res
````

就是 [Post Fix](https://zh.wikipedia.org/wiki/%E9%80%86%E6%B3%A2%E5%85%B0%E8%A1%A8%E7%A4%BA%E6%B3%95)

把東西放進stk 遇到operator pop兩次計算後放回stk


````
import operator
import math
````

````
## 480 ms, 13.4 MB
## 花太多時間了
def evalRPN(tokens):
    ops = {
        '+' : operator.add,
        '-' : operator.sub,
        '*' : operator.mul,
        '/' : None
          }
    if not tokens: return
    stk = []
    while tokens:
        cur = tokens[0]
        print('Current:', cur)
        if cur not in ops:
            stk.append(int(cur))
        elif cur in ops:
            if not stk: return
            num2 = stk.pop()
            num1 = stk.pop()
            if cur == '/':
                res = math.trunc(num1 / num2)
            else:
                res = ops[cur](num1, num2)
            stk.append(res)
        else: return
        print(stk)
        tokens = tokens[1:]
    return stk.pop()
````

````
# 超快的解法
## 根本不需要dict 簡單的對應就是了
## *, +, - 直接跟stk[-1]計算 可以少pop一次
def evalRPN(self, tokens):
    s = []
    for token in tokens:
        if token is '*':
            top2 = s.pop()
            s[-1] *= top2
        elif token is '/':
            top2 = s.pop()
            top1 = s.pop()
            s.append(int(top1 / top2)) #直接把float變成int 去掉小數點後的東西
        elif token is '+':
            top2 = s.pop()
            s[-1] += top2
        elif token is '-':
            top2 = s.pop()
            s[-1] -= top2
        else:
            s.append(int(token)) #去掉小數點
    return s.pop()
````

boolean DFS(Node cur, Node target, Set<Node> visited) {
    return true if cur is target;
    for (next : each neighbor of cur) {
        if (next is not in visited) {
            add next to visted;
            return true if DFS(next, target, visited) == true;
        }
    }

````
def dfs_itr(graph, start):
    visited, stack = set(), [start]
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            visited.add(vertex)
            stack.extend(graph[vertex] - visited) # extend: append another list
    return visited
````

````
def dfs_rec(graph, start, visited=None):
    if visited is None:
        visited = set() # set 操作起來比較快
    visited.add(start)
    for nxt in graph[start] - visited:
        dfs_rec(graph, nxt, visited)
    return visited
````

# Number of Islands

````
def numIslands(self, grid: List[List[str]]) -> int:
    num_islands = 0
    if grid:
        for r in range(len(grid)):
            for c in range(len(grid[0])):
                if grid[r][c] == '1':
                    island = [[r, c]]
                    while island:
                        [x, y] = island.pop()
                        grid[x][y] = '0'
                        if x+1 < len(grid) and grid[x+1][y] == '1':
                            island.append([x+1, y])
                        if x-1 >= 0 and grid[x-1][y] == '1':
                            island.append([x-1, y])
                        if y+1 < len(grid[0]) and grid[x][y+1] == '1':
                            island.append([x, y+1])
                        if y-1 >= 0 and grid[x][y-1] == '1':
                            island.append([x, y-1])
                    num_islands += 1
    return num_islands
````

# [Implement Queue using Stacks](https://leetcode.com/explore/learn/card/queue-stack/239/conclusion/1386/)

用 List 但是一個不夠

in_stk: 普通的 stack

out_stk: 接住所有 in_stk丟出來的東西 => 反向的 List

in_stk = [0, 1, 2, 3]

out_stk = [3, 2, 1, 0]


````
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.in_stk = []
        self.out_stk = []

    def push(self, x: int) -> None:
        """
        Push element x to the back of queue.
        """
        self.in_stk.append(x)

    def pop(self) -> int:
        """
        Removes the element from in front of queue and returns that element.
        """
        if not self.out_stk:
            while self.in_stk:
                self.out_stk.append(self.in_stk.pop())
        return self.out_stk.pop()

    def peek(self) -> int:
        """
        Get the front element.
        """
        if not self.out_stk:
            while self.in_stk:
                self.out_stk.append(self.in_stk.pop())
        return self.out_stk[-1]

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        return not self.in_stk and not self.out_stk
````

# [Flood Fill](https://leetcode.com/explore/learn/card/queue-stack/239/conclusion/1393/)

[Flood Fill 是三小](https://zh.wikipedia.org/wiki/Flood_fill)

aka. 小畫家油漆桶工具

## 題目敘述

圖片是2d矩陣 每個值代表一個pixel

給定座標(sr, sc)作為起始pixel 還有newColor是新的顏色

將起始pixel及相鄰四個方向 同顏色的其他pixel 塗上新的顏色 以此類推

return 塗色後的圖片

4 direction: +1 -1

````
# DFS + recursive
## 60 ms, 13.2 MB 意外的挺好的
def floodFill(image, sr, sc, newColor):
    sc_max = len(image[0]) - 1
    sr_max = len(image) - 1
    pre_color = image[sr][sc]
    if pre_color == newColor: return image # 避免無線迴圈
    image[sr][sc] = newColor
    if (sr > 0 and image[sr-1][sc] == pre_color):
        floodFill(image, sr-1, sc, newColor)
    if (sc > 0 and image[sr][sc-1] == pre_color):
        floodFill(image, sr, sc-1, newColor)
    if (sr < sr_max and image[sr+1][sc] == pre_color):
        floodFill(image, sr+1, sc, newColor)
    if (sc < sc_max and image[sr][sc+1] == pre_color):
        floodFill(image, sr, sc+1, newColor)
    return image
````

````
# 一樣的東西 更帥的寫法

def floodFill(self, image, sr, sc, newColor):
        rows, cols, orig_color = len(image), len(image[0]), image[sr][sc]
        
        def traverse(row, col):
            if (not (0 <= row < rows and 0 <= col < cols)) or image[row][col] != orig_color:
                return # 超過邊界或不同顏色 免做
            image[row][col] = newColor
            [traverse(row + x, col + y) for (x, y) in ((0, 1), (1, 0), (0, -1), (-1, 0))]
            # 沒有要存成 List 也做 List Comprehensive
        
        if orig_color != newColor: # 一樣顏色就免做了
            traverse(sr, sc)
            
        return image
````

````
# 44 ms
## 還是 DFS 少傳遞東西 速度自然會更快
def floodFill(self, image: List[List[int]], sr: int, sc: int, newColor: int) -> List[List[int]]:
    
    rs, cs = len(image), len(image[0])
    color = image[sr][sc]
    if color == newColor: return image

    def dfs(r, c):
        if image[r][c] == color:
            image[r][c] = newColor
            for nr,nc in ((r-1,c),(r+1,c),(r,c-1),(r,c+1)):
                if 0<=nr<rs and 0<=nc<cs: 
                    dfs(nr, nc)
    dfs(sr, sc)
    return image
````

````
def BFS(graph, root):
    visited, queue = set(), collections.deque([root])
    while queue:
        vertex = queue.popleft()
        for neightbour in graph[vertex]:
            if neighbour in graph not in visited:
                visited.add(neghbour)
                queue.append(neghbour)
````

````
# 快一倍 316 ms 的解法
## 基本上還是 BFS 亮點在於用第二個Queue儲存還沒被走到的點 所以不會有重複遍歷的問題
def updateMatrix(self, matrix: List[List[int]]) -> List[List[int]]:
        M = len(matrix)
        if M == 0: return [[]]
        N = len(matrix[0])
        cells = []
        for i, row in enumerate(matrix): # 還是不太會用這個
            for j, c in enumerate(row):
                if c == 1: # 相反 是把1放進Queue中
                    cells.append((i, j)) # Queue
        level = 1
        def is_adjacent(i, j): # 邊界測試跟
            if (((j==0) or matrix[i][j-1] >= level) and
                ((j==N-1) or matrix[i][j+1] >= level) and
                ((i==0) or matrix[i-1][j] >= level) and
                ((i==M-1) or matrix[i+1][j] >= level) 
               ):
                return False
            return True
        
        while cells:
            new_set = []
            for cell in cells:
                if not is_adjacent(cell[0],cell[1]):
                    new_set.append(cell) # 第二個Queue 存放還沒被走到的點
                    matrix[cell[0]][cell[1]] = level + 1 # 被走到了
            cells = new_set
            level += 1 # 往下一層
        return matrix
````
