# [1. Two Sum](https://leetcode.com/problems/two-sum/)

只到可以相加 = 目標的兩個數字

回傳它們的 index

### 思路1: 暴力破解 21%

第一個 Loop: 與目標相減 得到新目標

第二個 Loop: 找到新目標

時間複雜度: $O(N^2)$


```
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        if len(nums) <= 1:
            return None
        for a in range(len(nums)):
            temp_target = target - nums[a]
            for b in range(a+1, len(nums)):
                if nums[b] == temp_target:
                    return [a, b]
        return None
```

### 思路2: Dict 97%

把 Dict 當作 Hash Table 來用

dict 的索引(key): target - 某數字

dict 的值(value): 某數字的座標

```
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        if len(nums) <= 1: return None
        dict_hash = {}
        for i in range(len(nums)):
            if nums[i] in dict_hash:
                return [dict_hash[nums[i]], i]
            else:
                dict_hash[target - nums[i]] = i
```

# [21. Merge Two Sorted Lists](https://leetcode.com/problems/merge-two-sorted-lists/)

將兩個 List 合成一個並從小排到大

### 思路1: 遞迴 86%
比大小, 把小的放到第三個 list 中, 然後往下走

```
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        if not l1 and not l2:
            return
        elif not l1:
            return l2
        elif not l2:
            return l1
        if l1.val <= l2.val:
            head = l1
            l1 = l1.next
        else:
            head = l2
            l2 = l2.next
        
        head.next = self.mergeTwoLists(l1, l2)
        return head
```

### 思路2: 迴圈 98%

要記得回傳最後剩下的東西


```
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        p = ListNode(-1)
        head = p
        while l1 and l2:
            if l1.val <= l2.val:
                p.next, l1 = l1, l1.next
            else:
                p.next, l2 = l2, l2.next
            p = p.next
        
        if l1: p.next = l1
        else: p.next = l2
        
        return head.next
```
