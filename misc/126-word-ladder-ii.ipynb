## [127\. Word Ladder](https://leetcode.com/problems/word-ladder/)

Difficulty: **Medium**  

Related Topics: [Breadth-first Search](https://leetcode.com/tag/breadth-first-search/)


Given two words (_beginWord_ and _endWord_), and a dictionary's word list, find the length of shortest transformation sequence from _beginWord_ to _endWord_, such that:

1.  Only one letter can be changed at a time.
2.  Each transformed word must exist in the word list.

**Note:**

*   Return 0 if there is no such transformation sequence.
*   All words have the same length.
*   All words contain only lowercase alphabetic characters.
*   You may assume no duplicates in the word list.
*   You may assume _beginWord_ and _endWord_ are non-empty and are not the same.

**Example 1:**

```
Input:
beginWord = "hit",
endWord = "cog",
wordList = ["hot","dot","dog","lot","log","cog"]

Output: 5

Explanation: As one shortest transformation is "hit" -> "hot" -> "dot" -> "dog" -> "cog",
return its length 5.
```

**Example 2:**

```
Input:
beginWord = "hit"
endWord = "cog"
wordList = ["hot","dot","dog","lot","log"]

Output: 0

Explanation: The endWord "cog" is not in wordList, therefore no possible transformation.
```

## BFS



````python
from collections import deque, defaultdict

class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        wordList = set(wordList)
        if endWord not in wordList or not wordList:
            return 0
        queue = deque([beginWord])
        step = 0
        N = len(beginWord)
        seen = set()
        chars = set('abcdefghijklmnopqrstuvwxyz')
        
        while queue:
            step += 1
            for i in range(len(queue)):
                word = queue.popleft()
                for i in range(N):
                    for c in chars:
                        nextWord = word[:i] + c + word[i+1:]
                        if nextWord == endWord:
                            return step + 1
                        if nextWord not in seen and nextWord in wordList:
                            seen.add(nextWord)
                            queue.append(nextWord)

        return 0
````

### [126\. Word Ladder II](https://leetcode.com/problems/word-ladder-ii/)

```

beginWord = "hit",

[
  ["hit","hot","dot","dog","cog"],

```

````python
# Two Directions BFS
from collections import defaultdict
class Solution:
    def findLadders(self, beginWord: str, endWord: str, wordList: List[str]) -> List[List[str]]:
        wordList = set(wordList)
        direction = 1
        
        if endWord not in wordList:
            return []
        head, tail, res = {beginWord}, {endWord}, []
        chars = set('abcdefghijklmnopqrstuvwxyz')
        getNext = defaultdict(set)
        
        def dfs(currWord, path):
            if currWord == endWord:
                res.append(path)
            else:
                for nextWord in getNext[currWord]:
                    dfs(nextWord, path + [nextWord])
        
        while head and tail:
            if len(head) > len(tail):
                head, tail = tail, head
                direction *= -1
            s = set()
            wordList -= head
            while head:
                currWord = head.pop()
                for i in range(len(currWord)):
                    for c in chars:
                        nextWord = currWord[:i] + c + currWord[i+1:]
                        if nextWord in wordList:
                            s.add(nextWord)
                            if direction == 1:
                                getNext[currWord].add(nextWord)
                            else:
                                getNext[nextWord].add(currWord)
            head = s
            if s & tail:
                dfs(beginWord, [beginWord])
                return res
        return []
````
