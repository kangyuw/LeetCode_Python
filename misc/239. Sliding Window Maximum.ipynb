### [239\. Sliding Window Maximum](https://leetcode.com/problems/sliding-window-maximum/)

Difficulty: **Hard**  

Related Topics: [Heap](https://leetcode.com/tag/heap/), [Sliding Window](https://leetcode.com/tag/sliding-window/)


You are given an array of integersÂ `nums`, there is a sliding window of size `k` which is moving from the very left of the array to the very right. You can only see the `k` numbers in the window. Each time the sliding window moves right by one position.

Return _the max sliding window_.

**Example 1:**

```
Input: nums = [1,3,-1,-3,5,3,6,7], k = 3
Output: [3,3,5,5,6,7]
Explanation: 
Window position                Max
---------------               -----
[1  3  -1] -3  5  3  6  7       3
 1 [3  -1  -3] 5  3  6  7       3
 1  3 [-1  -3  5] 3  6  7       5
 1  3  -1 [-3  5  3] 6  7       5
 1  3  -1  -3 [5  3  6] 7       6
 1  3  -1  -3  5 [3  6  7]      7
```

**Example 2:**

```
Input: nums = [1], k = 1
Output: [1]
```

**Example 3:**

```
Input: nums = [1,-1], k = 1
Output: [1,-1]
```

**Example 4:**

```
Input: nums = [9,11], k = 2
Output: [11]
```

**Example 5:**

```
Input: nums = [4,-2], k = 2
Output: [4]
```

**Constraints:**

*   `1 <= nums.length <= 10<sup>5</sup>`
*   `-10<sup>4</sup> <= nums[i] <= 10<sup>4</sup>`
*   `1 <= k <= nums.length`

### Deque

1. iterative nums, push the element inside the queue

2. whenever push a new element, popout all smaller element inside: clear_queue()

3. the maximum element within the window is the top element of the queue

4. Advanced index, if the front index < current index, pop it out

Instead of storing value, storing index would be easier

````python
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        if k == 1:
            return nums
        if k > len(nums):
            return max(nums)
        
        def clear_queue(new):
            # have to start from right, since there might be smaller element behind max
            while queue and queue[-1] < new:
                queue.pop()
        
        queue = deque()
        ans = []
              
        for i in range(len(nums)):
            clear_queue(nums[i])
            queue.append(nums[i])
            if (i-k+1 >= 0):
                ans.append(queue[0])
                # if the max element is about to leave the window
                if (nums[i-k+1] == queue[0]):
                    queue.popleft()
            
        return ans
````

````python
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        
        ans = []
        queue = deque()
        
        for i in range(len(nums)):
            # the max index leave the window
            if queue and i - queue[0] == k:
                queue.popleft()
            
            # clear smaller index in queue
            while queue and nums[i] >= nums[queue[-1]]:
                queue.pop()
            queue.append(i)
            
            if i >= k-1:
                ans.append(nums[queue[0]])
        
        return ans
````
