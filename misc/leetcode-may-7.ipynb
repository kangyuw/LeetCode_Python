# Cousins in Binary Tree

在同一層 不同父節點的兩個點

## Solution 1: BST

Level Order Traversal

BST => Queue

```
class Solution:
    def isCousins(self, root: TreeNode, x: int, y: int) -> bool:
        def search(root, target):
            queue = [[root, 0, None]]
            while queue != None:
                node, depth, parent = queue.pop(0)
                if node.val == target:
                    return depth, parent
                if node.left:
                    queue.append([node.left, depth+1, node.val])
                if node.right:
                    queue.append([node.right, depth+1, node.val])
        depth_x, parent_x = search(root, x)
        depth_y, parent_y = search(root, y)
        
        return (depth_x==depth_y) and (parent_x != parent_y)
        
```

## Solution 2: Lookup

dfs 配合 dict


```
class Solution(object):
    def isCousins(self, root, x, y):
       
        lookup = {} #dictionary for maintaining the level and parent of the node
        def dfs(root, level=0, parent=None):
            if root:
                if root.val in (x, y): 
                    lookup[root.val] = (level, parent)
                dfs(root.left, level=level+1, parent=root.val)
                dfs(root.right, level=level+1, parent=root.val)
        dfs(root)
        return lookup[x][0] == lookup[y][0] and lookup[x][1] != lookup[y][1]
```
