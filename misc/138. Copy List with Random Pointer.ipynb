### [138\. Copy List with Random Pointer](https://leetcode.com/problems/copy-list-with-random-pointer/)

Difficulty: **Medium**


A linked list is given such that each node contains an additional random pointer which could point to any node in the list or null.

Return a of the list.

The Linked List is represented in the input/output as a list of `n` nodes. Each node is represented as a pair of `[val, random_index]` where:

*   `val`: an integer representing `Node.val`
*   `random_index`: the index of the node (range from `0` to `n-1`) where random pointer points to, or `null` if it does not point to any node.

*   Number of Nodes will not exceed 1000.

一般的 Link List:

+ Time: O(N)

+ Space: O(1)

一般而言, 複製 Link List 需要一個 pointer 往前遍歷, 問題在於什麼時候要複製 Random

### Linear Space Solution

使用 Hash Table

2 Pass: 只複製 Nodes + 連線

複製 Node 存進 Hash Table

連線: next & random 使用 hash table 的索引


````python
class Solution:
    def copyRandomList(self, head: 'Node') -> 'Node':
        
        dic = dict() # hash table
        p = q = head
        
        # copy nodes
        while p:
            dic[p] = Node(p.val)
            p = p.next
        
        # wired
        while q:
            dic[q].next = dic.get(q.next)
            dic[q].random = dic.get(q.random)
            q = q.next
        
        return dic.get(head)
````

### Constant Space Solution
[參考文章](https://leetcode.com/problems/copy-list-with-random-pointer/discuss/43491/A-solution-with-constant-space-complexity-O(1)-and-linear-time-complexity-O(N))

3 Pass:

1. 遍歷原本的 list, 複製每一個 node 黏在原本的後面

2. 遍歷新的 list 將 random 連起來

3. 恢復原本的 list 將複製的 node 取出來


````python
class Solution:
    def copyRandomList(self, head: 'Node') -> 'Node':
        new = head
        
        # 複製原本的 node 然後黏在它的後面
        while new:
            copy = Node(new.val)
            copy.next = new.next
            new.next = copy
            new = copy.next

        # 連線 random
        ptr = head
        while ptr:
            ptr.next.random = ptr.random and ptr.random.next
            ptr = ptr.next.next

        # 取出複製
        ptr = head
        copy = head_copy = head and head.next

        while ptr:
            ptr.next = ptr = copy.next
            copy.next = copy = ptr and ptr.next

        return head_copy
````

### Faster Solution

````python
class Solution:
    def copyRandomList(self, head: 'Node') -> 'Node':
        if not head:
            return None
        
        dic = {}
        curr = head
        
        while curr:
            dic[curr] = Node(curr.val)
            curr = curr.next
        
        curr = head
        while curr:
            if curr.random:
                dic[curr].random = dic[curr.random]
            if curr.next:
                dic[curr].next = dic[curr.next]
            curr = curr.next
        
        return dic[head]
````
