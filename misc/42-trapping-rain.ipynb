### [42\. Trapping Rain Water](https://leetcode.com/problems/trapping-rain-water/)

Difficulty: **Hard**


Given _n_ non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it is able to trap after raining.

![](https://assets.leetcode.com/uploads/2018/10/22/rainwatertrap.png)  
<small style="display: inline;">The above elevation map is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped. **Thanks Marcos** for contributing this image!</small>

**Example:**

```
Input: [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
```

## Solution 1: 2 Pointer

86.5%

### Head
+ 向右邊找到 `height[cur] >= height[Head]`
+ 計算容積 = `sum(height[cur] - height[i]) for i in range(Head, cur)`
+ Head = cur

Tail: 向左

直到兩個相會為止

````
class Solution:
    def trap(self, height):
        n = len(height)
        if n == 1 or n == 0: return 0
        head, tail = 0, n - 1
        temp_r, temp_l, total = 0, 0, 0
        while head < tail:
            for i in range(head + 1, n):
                if height[head] > height[i]:
                    temp_r += (height[head] - height[i])
                else:
                    total += temp_r
                    temp_r = 0
                    head = i
                    break
            if head >= tail: break
            for j in range(tail - 1, -1, -1):
                if height[tail] > height[j]:
                    temp_l += (height[tail] - height[j])
                else:
                    total += temp_l
                    temp_l = 0
                    tail = j
                    break
        return total
````

## Faster

不需要每次都移動 head, tail 只需要移動比較小的那邊就好

````
class Solution:
    def trap(self, height: List[int]) -> int:
        
        if not height:
            return 0
        
        l = 0
        r = len(height) - 1
        t = 0
        max_l = height[l]
        max_r = height[r]
        
        while l < r:
            if max_l < max_r:
                l += 1
                max_l = max(max_l, height[l])
                t += max_l - height[l]
                
            else:
                r -= 1
                max_r = max(max_r, height[r])
                t += max_r - height[r]
        return t
````
