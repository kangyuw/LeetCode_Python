# 4. Median of Two Sorted Arrays

[Reference](https://leetcode.com/problems/median-of-two-sorted-arrays/discuss/2481/Share-my-O(log(min(mn)))-solution-with-explanation)

尋找兩個矩陣的中位數

暴力排序: $O(log(m+n))$

## [Binary Search](https://www.youtube.com/watch?v=LPFhl65R7ww)

$O(log(min(m,n)))$

將矩陣分為滿足兩個條件的兩個部分

X 分為 X_left X_right

Y 分為 Y_left Y_right

```
1. len(left) == len(right)

2. max(left) <= min(right)
```

滿足條件後

$ 中位數 = avg(max(left) + min(right)) $

切到某部分空了: $ -\infty \ , \ \infty  $

### 條件1
左右兩邊的個數相等
```
Even:
ptX + ptY = (nX + nY) // 2
Odd:
ptX + ptY = (nX + nY + 1) // 2
```

### 條件2:
X 左邊最大值 < Y 右邊最小值

Y 左邊最大值 < X 右邊最小值
```

````
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        x, y = len(nums1), len(nums2)
        
        if x > y:
            nums1, nums2, x, y = nums2, nums1, y, x
        
        if x == 0:
            if y % 2 != 0:
                return nums2[y//2]
            else:
                return (nums2[y//2-1] + nums2[y//2])/2
        
        start = 0
        end = x
        while start <= end:
            
            ptX = (start+end) // 2
            maxLeftX = nums1[ptX-1] if ptX != 0 else -float('inf')
            minRightX = nums1[ptX] if ptX != x else float('inf')
            
            ptY = (x + y + 1) // 2 - ptX
            
            print(ptX, ptY)
            
            maxLeftY = nums2[ptY-1] if ptY != 0 else -float('inf')
            minRightY = nums2[ptY] if ptY != y else float('inf')
            
            if (maxLeftX <= minRightY and maxLeftY <= minRightX):
                if (x + y)%2 == 0: # 偶數
                    return (max(maxLeftX, maxLeftY) + min(minRightX, minRightY)) / 2
                else:
                    return max(maxLeftX, maxLeftY)
                
            elif (maxLeftX > minRightY):
                end = ptX - 1
            
            else:
                start = ptX + 1
````
