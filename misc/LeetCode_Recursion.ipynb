## [Reverse String](https://leetcode.com/explore/featured/card/recursion-i/250/principle-of-recursion/1440/)

這次用遞迴的方式

Base Case: tail < head

````
def reverseString(s):
    
    def helper(s, head, tail):
        if tail <= head:
            return s
        s[head], s[tail] = s[tail], s[head]
        return helper(s, head +1, tail -1)
    helper(s, 0, len(s) - 1)
    print(s)
````

## [Swap Nodes in Pairs](https://leetcode.com/explore/featured/card/recursion-i/250/principle-of-recursion/1681/)

把兩個相鄰的元素交換

1. 交換第一個跟第二個
2. 把第三個丟進去

Base Case: 走到最後一個點

````
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
````

````
def toLinkList(lst):
    for idx, val in enumerate(lst):
        if idx == 0:
            head = ListNode(val)
            pre = head
        else:
            pre.next = ListNode(val)
            pre = pre.next
        print(pre.val)
        if idx == len(lst) - 1: return head
````

````
LinkList = toLinkList([1,2,3,4])
````

````
def swapPairs(head):
    if not head or not head.next:
        return head
    sec = head.next
    head.next = swapPairs(sec.next)
    sec.next = head
    return sec
````

````
def swapPairs_iter(head):
    if not head or not head.next:
        return head
    fst, sec = head, head.next
    head, pre = sec, None
    while fst and sec:
        if pre:
            pre.next = sec
        fst.next = sec.next
        sec.next = fst
        pre = fst
        fst = fst.next
        sec = fst.next if fst else None
    return head
````

## [Pascal's Triangle](https://leetcode.com/explore/featured/card/recursion-i/251/scenario-i-recurrence-relation/1659/)

從最底下開始 一層一層走上去

撞到0之後 開始往下計算

for i-th row and j-th column

F (i, j) = F (i-1, j-1) + F (i-1, j)

Base Case: row 0: [], row 1: [[1]]

````
def generate(numRows):
    if numRows == 0: return []
    elif numRows == 1: return [[1]]
    res = generate(numRows -1)
    row = []
    for idx in range(numRows):
        if idx == 0 or idx == numRows - 1:
            row.append(1)
        else:
            row.append(res[numRows-2][idx-1] + res[numRows-2][idx])
    res.append(row)
    return res
````

## [Pascal's Triangle II](https://leetcode.com/explore/featured/card/recursion-i/251/scenario-i-recurrence-relation/1660/)

上一題回傳當下那行就好

Base Case: row 0: [1], row 1: [1,1]

````
def getRow(self, rowIndex: int) -> List[int]:
    if rowIndex == 0: return [1]
    elif rowIndex == 1: return [1,1]
    pre = self.getRow(rowIndex -1)
    return [1] + [pre[idx] + pre[idx+1] for idx in range(len(pre)-1)] + [1]
````

## [Fibonacci Number](https://leetcode.com/explore/featured/card/recursion-i/255/recursion-memoization/1661/)

計算該數列時 會重複計算到不少次

利用 dict 的特性儲存結果 方便往後調用

Base Case: {0: 0, 1: 1}

````
## top-down using memoization
## Time: O(N), Space: O(N)
class Solution:
    def fib(self, N):
        if N <= 1: return N
        self.cache = {0: 0, 1: 1}
        return self.fib_memo(N)
    
    def fib_memo(self, N):
        if N in self.cache.keys():
            return self.cache[N]
        self.cache[N] = self.fib_memo(N-1) + self.fib_memo(N-2)
        return self.fib_memo(N)
````

````
## Iterative top-down
## Time: O(N), Space: O(1)
def fib(N):
    if N<=1: return N
    prev1 = 0
    prev2 = 1
    k = 2
    while k <= N:
        res = prev1 + prev2
        prev1 = prev2
        prev2 = res
        k += 1
    return res
````

````
fib(51)
````

## [Climb Stairs](https://leetcode.com/explore/featured/card/recursion-i/255/recursion-memoization/1662/)

只能走1步或2步

climbStairs(X) = climbStairs(X - 1) + climbStairs(X - 2)

cache = {1:1, 2:2}

Base Case: {1: 1, 2: 2}

````
## Recursion with Memoization (top-down)
class Solution:
    def climbStairs(self, N):
        if N <= 2: return N
        self.cache = {1: 1, 2: 2}
        return self.memo(N)
    
    def memo(self, N):
        if N in self.cache.keys():
            return self.cache[N]
        self.cache[N] = self.memo(N-1) + self.memo(N-2)
        return self.memo(N)
````

## Top-Down vs Bottom-up

### Top-Down
從第n階向下思考 要爬到第n階 有多少種可能
1. 在 n-1 階 再爬1層
2. 在 n-2 階 再爬2層

1. 到達 n-1 層有幾種可能
2. 到達 n-2 層有幾種可能

### Bottom-up
向上思考 只要知道當前所在的層數, 它的 n-1 及 n-2 各有多少可能



````
## Bottom-up
def climbStaris_BU(n):
    if n==1: return 1
    a, b = 1, 2
    for _ in range(2,n):
        a, b = b, a + b
    return b
````

## [Maximum Depth of Binary Tree](https://leetcode.com/explore/learn/card/recursion-i/256/complexity-analysis/2375/)

base case: root 不存在了 回傳走了多少步

````
def maxDepth(self, root):
    if not root: return 0
    return max(self.maxDepth(root.left), self.maxDepth(root.right)) + 1
````

````
# 快上不少
def maxDepth_tail(self, root):
    def helper(root, curDepth):
        if root is None: return curDepth - 1
        else:
            return max(helper(root.left, curDepth), helper(root.right, curDepth))
    return helper(root, 1)
````

##  [Pow(x, n)](https://leetcode.com/explore/learn/card/recursion-i/256/complexity-analysis/2380/)

求得 x 的 n 次方

Base Case: if n == 0: return 1

遞迴: 將 x 折半 一直做到 0

三種情況:

1. n 是偶數, 直接除以二

    ex. 2^100 => 2^50 * 2^50

2. n 是奇數, 除以二後多乘一次
    
    ex. 2^25 => 2^12 * 2^12

3. n 是負數, 取倒數

    ex. 1.0 / Pow(x, -n)

````
class Solution(object):
    def pow(self, x, n):
        if n == 0: return 1
        if n < 0:
            return 1.0 / self.helper(x, -n)
        else:
            return self.helper(x, n)
        
    def helper(self, x, n):
        if n == 0: return 1
        half = self.helper(x, n//2)
        if n % 2 == 0:
            return half * half
        else:
            return half * half * x 
````

## [Merge Two Sorted Lists](https://leetcode.com/explore/learn/card/recursion-i/253/conclusion/2382/)

Base Case: 某個 List 已經沒東西了 return 剩下的元素

比較 L1 跟 L2 的當前節點

把較小的取出來 當前節點往下走

````
l1 = toLinkList([1,2,4])
l2 = toLinkList([1,3,4])
````

````
def mergeTwoLists(l1, l2):
    if not l1 and not l2:
        return
    elif not l1: return l2
    elif not l2: return l1

    if l1.val < l2.val:
        head = l1
        l1 = l1.next
    else:
        head = l2
        l2 = l2.next

    head.next = mergeTwoLists(l1, l2)
    return head
````

````
# 更簡潔的寫法
def mergeTwoLists_brief(l1, l2):
    if not l1 or not l2:
        return l1 or l2
    
    if l1.val < l2.val:
        l1.next = mergeTwoLists(l1.next, l2)
        return l1
    else:
        l2.next = mergeTwoLists(l1, l2.next)
        return l2
````

````
# in-place, iteratively
def mergeTwoLists_iter(l1, l2):
    dummy = cur = ListNode(0)
    while l1 and l2:
        if l1.val < l2.val:
            cur.next = l1
            l1 = l1.next
        else:
            cur.next = l2
            l2 = l2.next
        cur = cur.next
    cur.next = l1 or l2
    return dummy.next
````

## [K-th Symbol in Grammar](https://leetcode.com/explore/learn/card/recursion-i/253/conclusion/1675/)

根據上一行的數字 0 產生 01, 1 產生 10

row1: 0

row2: 01

row3: 0110

row4: 01101001

求得第 N 行 第 K 個數字 (從1開始)

Base Case: row1 [0]

### 思路1

類似帕斯卡 => 太慢 沒必要計算每一行 O(2^N)

### 思路2

只要找到路徑上的東西就夠了

路徑: 看上一行 index // 2 是什麼東西就好

偶數回傳一樣的 奇數回傳不一樣的 => XOR

````
## TLE 沒有必要一個一個產生
## 只需要知道連到 Base Case 路徑上的東西就夠了
class Solution(object):
    def kthGrammar(self, N, K):
        return self.getRow(N)[K - 1]
    
    
    def getRow(self, N):
        if N == 1: return [0]
        pre = self.getRow(N-1)
        res = []
        for digit in pre:
            if digit == 1:
                res.extend([1,0])
            else:
                res.extend([0,1])
        return res
````

````
def kthGrammar(N, K):
    def helper(N, index):
        if N == 1: return 0
        if index % 2 == 0:
            return helper(N-1, index // 2)
        else:
            return 1 if helper(N-1, index // 2) == 0 else 0
    return helper(N, K-1)
````

````
def kthGrammar_iter(N,K):
    res = 0
    while K > 1:
        K = K + 1 if K % 2 else K / 2
        res ^= 1 ## ixor(res, 1)
    return res
````

````
def kthGrammar_recur(N, K):
    if N == 1: return 0
    if K <= 2**(N-2):
        return kthGrammar(N-1,K)
    else:
        return 1^kthGrammar(N-1,K-2**(N-2)) ## xor
````

## [Unique Binary Search Trees II](https://leetcode.com/explore/learn/card/recursion-i/253/conclusion/2384/)

輸出從 1 ... N 節點的樹, 有幾種不同的獨特BST


```
Input: 3

Output: 
[
  [1,null,3,2],
  [3,2,null,1],
  [3,1,null,null,2],
  [2,1,3],
  [1,null,2,null,3]
]

```

Pre-Order Traversal: VLR

Base Case: 只剩下一個點

BST: 左小右大

遍歷選擇點作為 root 比較小的丟左邊 比較大的丟右邊


````
class Solution(object):
    def generateTrees(self, n):
        """
        :type n: int
        :rtype: List[TreeNode]
        """
        if n <= 0: return []
        return self.helper(1,n)
    
    
    def helper(self, start, end):
        res = []
        for cur in range(start, end + 1):
            for left_node in self.helper(start, cur-1):
                for right_node in self.helper(cur+1, end):
                    root = TreeNode(cur)
                    root.left = left_node
                    root.right = right_node
                    res.append(root)
        return res or [None]
````

````
## 稍快的方法 用 Hash Table 儲存結果 一次調用
def generateTrees(self, n):
    """
    :type n: int
    :rtype: List[TreeNode]
    """
    d = {}
    def generateTreeUtil(start, end):
        if start > end:
            return [None]
        key = (start, end)
        if key not in d:
            result = []
            for i in range(start, end + 1):
                for l in generateTreeUtil(start, i - 1):
                    for r in generateTreeUtil(i + 1, end):
                        newRoot = TreeNode(i)
                        newRoot.left = l
                        newRoot.right = r
                        result.append(newRoot)
            d[key] = result
        return d[key]
    return generateTreeUtil(1, n) if n else []
````
