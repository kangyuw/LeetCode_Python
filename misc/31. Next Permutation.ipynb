# Generate All Permutation of A String
[video](https://www.youtube.com/watch?v=GCm7m5671Ps)
[46. Permutation](https://leetcode.com/problems/permutations/)

* "complete search" / "generate all" is a backtracking problem

* Brute Force: O(n!)

## 3 Keys of Backtracking

### Choice

* choose a charcter from remaining charcters
* each decision is a recursion function

### Constrain

* every charcter can only be chosen once

### Goal

* generated list has same length as input ary

````python
def permutation(ary: List[]):
    # Recursive with each decision
    res = []
    # constrain 
    if len(ary) == 0:
        return res
    
    def backtracking(res, item, ary):
        # Reach the Goal
        if len(item) == len(ary):
            res.append(item)
            return
        for i in range(len(ary)):
            # Constrain: each element can only be chosen once
            if ary[i] in item:
                continue
            item.append(ary[i])
            backtracking(res, item.copy(), ary)
            # express all possiblility, kick out from decision pool
            item.pop()
    
    backtracking(res, [], ary)
    return res
````

### [31\. Next Permutation](https://leetcode.com/problems/next-permutation/)

Difficulty: **Medium**  

Related Topics: [Array](https://leetcode.com/tag/array/)


Implement **next permutation**, which rearranges numbers into the lexicographically next greater permutation of numbers.

If such arrangement is not possible, it must rearrange it as the lowest possible order (ie, sorted in ascending order).

The replacement must be and use only constant extra memory.

Here are some examples. Inputs are in the left-hand column and its corresponding outputs are in the right-hand column.

`1,2,3` → `1,3,2`  
`3,2,1` → `1,2,3`  
`1,1,5` → `1,5,1`

### Solution: Brute Force

We do not see a pattern

Generate all possible permutation

time: O(n!)

### Solution: Backtracking

[video](https://youtu.be/quAS1iydq7U)

eg.
```
find the next permutation of:
[6,2,1,5,4,3,0]

find the longest decreasing part of array: [1, 5, 4, 3, 0]

Since 1 is the first element, 0 must run through all its permutation

Since 5 4 3 is the biggest permutation:
[6, 2, 1, 5, 4, 3, 0] the last permutation of 1

swap the position of 1 and 3:
[6, 2, 3, 5, 4, 1, 0] the last permutation of 3

reverse the last part:
[6, 2, 3, 0, 1, 4, 5] the first permutation of 3
```

time: O(n)

space: O(1)


````python
class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        We do not know the exact rule of permutation.
        Therefore, we should try every permutation and find out the correct one.
        However, it will have time complexity of O(n!)
        """
        n = len(nums)
        
        # find the longest non-increasing suffix
        right = n - 1
        while nums[right] <= nums[right - 1] and right - 1 >= 0:
            right -= 1
        
        if right == 0:
            return self.reverse(nums, 0, n-1)
        
        # find swap point
        swap = right - 1
        successor = 0
        # find rightmost successor
        for i in range(n-1, swap, -1):
            if nums[i] > nums[swap]:
                successor = i
                break
        # swap
        nums[swap], nums[successor] = nums[successor], nums[swap]
        # reverse suffix
        self.reverse(nums, swap + 1, n - 1)
        
    def reverse(self, nums, left, right):
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1
````
