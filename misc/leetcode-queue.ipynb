````
class MyCircularQueue:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the queue to be k.
        """
        self.data = [None] * k
        self.head = self.tail = -1
        self.size = k
        self.curSize = 0

    def enQueue(self, value: int) -> bool:
        """
        Insert an element into the circular queue. Return true if the operation is successful.
        """
        if self.isFull(): return False
        if self.isEmpty(): self.head  = 0
        self.tail = (self.tail + 1) % self.size
        self.data[self.tail] = value
        self.curSize += 1
        return True

    def deQueue(self) -> bool:
        """
        Delete an element from the circular queue. Return true if the operation is successful.
        """
        if self.isEmpty(): return False
        if self.head == self.tail:
            self.head = self.tail = -1
            self.curSize -= 1
            return True
        self.head = (self.head + 1) % self.size
        self.curSize -= 1
        return True

    def Front(self) -> int:
        """
        Get the front item from the queue.
        """
        return self.data[self.head] if not self.isEmpty() else -1
        

    def Rear(self) -> int:
        """
        Get the last item from the queue.
        """
        return self.data[self.tail] if not self.isEmpty() else -1

    def isEmpty(self) -> bool:
        """
        Checks whether the circular queue is empty or not.
        """
        return self.curSize == 0

    def isFull(self) -> bool:
        """
        Checks whether the circular queue is full or not.
        """
        return self.curSize == self.size
````

    while (queue is not empty) {
        step = step + 1;
        // iterate the nodes which are already in the queue
        int size = queue.size();
        for (int i = 0; i < size; ++i) {
            Node cur = the first node in queue;
            return step if cur is target;
            for (Node next : the neighbors of cur) {

# [Number of Islands](https://leetcode.com/explore/learn/card/queue-stack/231/practical-application-queue/1374/)

2d 網格地圖

1: 陸地

0: 水

計算島嶼的數量 四邊當作都是 0

島嶼: 被水隔離的陸地區塊, 只看垂直與水平方向

暴力法: 每經過一點 就檢查四周有沒有其他陸地 => 不需要BFS

從val = 1 的點開始做BFS

做到最後就給 num_island += 1

然後把visit過的點val設為0

ptr = 走到哪個點了

BFS: 把{x,+1 y}, {x, y+1} 放進 Queue中 並設為 '0'

DFS: 遞迴 對{x,+1 y}, {x, y+1}再做一次 直到碰到'0' 或是碰到底了

DFS快多了

````
# BFS
from collections import deque
class Solution:
    
    def numIslands(self, grid: List[List[str]]) -> int:
        num_island = 0
        if not grid or not grid[0]:
            return 0
        x_lim, y_lim = len(grid), len(grid[0])
        for x in range(x_lim):
            for y in range(y_lim):
                if grid[x][y] == '1':
                    self.BFS(grid, x, y)
                    num_island += 1
        return num_island
    
    def BFS(self, grid, x, y):
        queue = deque()
        queue.append((x, y))
        grid[x][y] = '0'
        while queue:
            x, y = queue.popleft()
            if self.check(grid, x+1, y):
                queue.append((x+1,y))
            if self.check(grid, x-1, y):
                queue.append((x-1, y))
            if self.check(grid, x, y+1):
                queue.append((x,y+1))
            if self.check(grid, x, y-1):
                queue.append((x,y-1))
        
    def check(self, grid, x, y):
        x_lim = len(grid)
        y_lim = len(grid[0])
        return  not (x >= x_lim or y >= y_lim or x < 0 or y < 0 or grid[x][y] == '0')
````

````
# DFS
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        num_islands = 0
        if grid:
            for r in range(len(grid)):
                for c in range(len(grid[0])):
                    if grid[r][c] == '1':
                        island = [[r, c]]
                        while island:
                            [x, y] = island.pop()
                            grid[x][y] = '0'
                            if x+1 < len(grid) and grid[x+1][y] == '1':
                                island.append([x+1, y])
                            if x-1 >= 0 and grid[x-1][y] == '1':
                                island.append([x-1, y])
                            if y+1 < len(grid[0]) and grid[x][y+1] == '1':
                                island.append([x, y+1])
                            if y-1 >= 0 and grid[x][y-1] == '1':
                                island.append([x, y-1])
                        num_islands += 1
        return num_islands
````

````
# MAGIC
def numIslands(self, grid):
    def sink(i, j):
        if 0 <= i < len(grid) and 0 <= j < len(grid[i]) and grid[i][j] == '1':
            grid[i][j] = '0'
            map(sink, (i+1, i-1, i, i), (j, j, j+1, j-1))
            return 1
        return 0
    return sum(sink(i, j) for i in range(len(grid)) for j in range(len(grid[i])))
````

````
class Solution:
    def openLock(self, deadends: List[str], target: str) -> int:
        if '0000' in deadends: return -1
        if '0000' == target: return 0
        dead = set(deadends)
        
        step = 0
        queue = collections.deque([('0000', 0)])
        
        while queue:
            cur_num, step = queue.popleft()
            for i in range(4):
                ## +1
                new_digit = int(cur_num[i]) + 1
                if new_digit > 9: new_digit = 0
                new_num = cur_num[:i] + str(new_digit) + cur_num[i+1:]
                
                if new_num == target: return step + 1
                if new_num not in dead:
                    queue.append((new_num, step + 1))
                    dead.add(new_num)
                    
                ## -1
                new_digit = int(cur_num[i]) - 1
                if new_digit < 0: new_digit = 9
                new_num = cur_num[:i] + str(new_digit) + cur_num[i+1:]
                
                if new_num == target: return step + 1
                if new_num not in dead:
                    queue.append((new_num, step + 1))
                    dead.add(new_num)
        return -1
````

````
## TLE
def numSquares(n: int) -> int:
    if n <= 3:
        return n
    
    q = collections.deque()
    base = 1
    step = 1

    while base * base <= n:
        node = base * base
        q.append((node, base, step))
        base += 1
        
    while q:
        node, base, step = q.popleft()
        if node == n: return step
        for i in range(1, base + 1):
            nxt_node = node + i * i
            if nxt_node <= n:
                q.append((nxt_node, i, step + 1))
````

````
## 差別在於一算出來就可以判定了　不用等他再從queue中跑出來
## 然而時間空間的表現頗差
## 應該可以少queue一個參數減少空間複雜度
class Solution:
    def numSquares(self, n: int) -> int:
        if n <= 3:
            return n
        

        q = collections.deque()
        base = 1
        step = 1

        while base * base <= n:
            node = base * base
            if node == n: return step
            q.append((node, base, step))
            base += 1

        while q:
            node, base, step = q.popleft()
            for i in range(1, base + 1):
                nxt_node = node + i * i
                if nxt_node <= n:
                    if nxt_node == n: return step + 1
                    q.append((nxt_node, i, step + 1))
````

````
## 很快的解法 四平方核定理

class Solution:
    def isSquare(self, n):
        r = int(n**0.5)
        return r*r == n
        
    def numSquares(self, n: int) -> int:
        # Legendre's four-square theorem
        # https://en.wikipedia.org/wiki/Lagrange%27s_four-square_theorem
        # 說明每個正整數均可表示為4個整數的平方和
        # ans <= 4
        # and 3 iff n is NOT of the form (4**a)*(8b+7)
    
        # check if answer = 1
        # 為某正整數的平方
        if self.isSquare(n):
            return 1
        
        # check if answer = 2
        # 為某正整數的平方
        i = 1
        while i*i < n:
            if self.isSquare(n - i*i):
                return 2
            i +=1
        
        # check if answer is NOT 3
        while n%4 == 0:
            n = n/4
        if n%8 == 7:
            return 4
        
        # answer must be 3 if none of the above
        return 3
````
