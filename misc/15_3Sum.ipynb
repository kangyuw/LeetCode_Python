### [15\. 3Sum](https://leetcode.com/problems/3sum/)

Difficulty: **Medium**


Given an array `nums` of _n_ integers, are there elements _a_, _b_, _c_ in `nums` such that _a_ + _b_ + _c_ = 0? Find all unique triplets in the array which gives the sum of zero.

找三個相加為零的組合

**Note:**

The solution set must not contain duplicate triplets.

**Example:**

```
Given array nums = [-1, 0, 1, 2, -1, -4],

A solution set is:
[
  [-1, 0, 1],
  [-1, -1, 2]
]
```

## 類似問題 Two Sum

找到相加為目標的組合 (只有一組)

````
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        d={}
        for ind,num in enumerate(nums):
            if target-num in d:
                return ind,d.get(target-num)
            d[num]=ind
````

## Solution 1. Two Pointers

### V1. 固定 head 與 tail, 移動 center

兩層迴圈: $O(N^2)$

1st loop:

    比較一開始的結果決定方向
2nd loop:

1. 總和 >= 0: 中間往左走, 直到
  
    1. 撞到 head

    2. 總和 < 0

    3. 總和 == 0

### 問題: 無法避免重複的結果

解決方式: 當 head 或 tail 移動時 跳過重複的 => 無法解決

````
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        head, tail = 0, len(nums) - 1
        res = []
        while(head <= tail - 2):
            center = (head + tail) // 2
            cur_target = nums[head] + nums[tail]
            cur_sum = cur_target + nums[center]
            
            if cur_sum >= 0:
                while (center > head):
                    cur_sum = cur_target + nums[center]
                    if cur_sum > 0:
                        center -= 1
                    elif cur_sum < 0:
                        break
                    else:
                        res.append([nums[head], nums[center], nums[tail]])
                        break
                if nums[tail] == nums[tail - 1]:
                    tail -=1
                tail -= 1
            
            elif cur_sum < 0:
                while (center < head):
                    cur_sum = cur_target + nums[center]
                    if cur_sum > 0:
                        break
                    elif cur_sum < 0:
                        center += 1
                    else:
                        res.append([nums[head], nums[center], nums[tail]])
                        break
                if nums[head] == nums[head+1]:
                    head += 1
                head += 1
        
        return res
````

### V2. 固定最左側, 移動 head, tail

86.86%

解決重複: head, tail 移動前判斷下一個是不是一樣的, 是就跳過

## 判斷出該怎麼解, 就盡量往那個做法靠攏

````
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        res = []
        
        for i in range(len(nums) - 2):
            if nums[i] > 0: return res
            if i > 0 and nums[i] == nums[i-1]: continue
            
            head = i + 1
            tail = len(nums) - 1
            
            while head < tail:
                cur_sum = nums[i] + nums[head] + nums[tail]
                if cur_sum < 0:
                    head += 1
                elif cur_sum > 0:
                    tail -= 1
                else:
                    res.append([nums[i], nums[head], nums[tail]])
                    
                    while head < tail and nums[head] == nums[head+1]:
                        head += 1
                    while head < tail and nums[tail] == nums[tail-1]:
                        tail -= 1
                    head += 1
                    tail -= 1
        
        return res
````
