2. 所有在左半的元素都比在右半的元素小

如何保證? max(leftX) <= min(rightY) and max(leftY) <= min(rightX)

之後就能得到中位數


Solution:

+ 將短的矩陣當作 X, 長的當作 Y

+ 如果 X 不存在, 直接回傳 Y 的中位數

binary search:

1. 從 X 的中點開始, 取 X\[midX-1] 及 X\[midX]

2. 為了滿足條件一: (midX + midY) * 2 = len(X) + len(Y)

    + midY = (len(X) + len(Y) + 1) // 2 - ptX, +1 是為了避免奇數被去掉

3. 如果滿足條件二: (maxLeftX <= minRightY and maxLeftY <= minRightX)

回傳中位數

    1. 總長度是偶數: (max(maxLeftX, maxLeftY) + min(minRightX, minRightY)) / 2
    2. 總長度是奇數: max(maxLeftX, maxLeftY)

4. 還沒找到中位數
    
    + 如果 leftX > rightY: leftX 太大了, hi = midX - 1
    + 如果 leftX < rightY: leftX 不夠大, lo = midX + 1

Time: O(log(min(x,y)))

````python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        x, y = len(nums1), len(nums2)
        
        if x > y:
            nums1, nums2, x, y = nums2, nums1, y, x
        
        if x == 0:
            if y % 2 != 0:
                return nums2[y//2]
            else:
                return (nums2[y//2-1] + nums2[y//2])/2
        
        lo = 0
        hi = x
        while lo <= hi:
            # start from mid
            midX = (lo+hi) // 2
            maxLeftX = nums1[midX-1] if midX != 0 else -float('inf')
            minRightX = nums1[midX] if midX != x else float('inf')
            
            # Condition 1 
            midY = (x + y + 1) // 2 - midX
            maxLeftY = nums2[midY-1] if midY != 0 else -float('inf')
            minRightY = nums2[midY] if midY != y else float('inf')
            
            # Condition 2
            if (maxLeftX <= minRightY and maxLeftY <= minRightX):
                if (x + y)%2 == 0:
                    return (max(maxLeftX, maxLeftY) + min(minRightX, minRightY)) / 2
                else:
                    return max(maxLeftX, maxLeftY)
            # leftX is still too big
            elif (maxLeftX > minRightY):
                hi = midX - 1
            # leftX is not big enough
            else:
                lo = midX + 1
````

[33\. Search in Rotated Sorted Array](https://leetcode.com/problems/search-in-rotated-sorted-array/)

從被翻轉的陣列中找特定數字

判斷 mid 與 target 之間的關係

如果 mid 較大, 開頭也較 target 大 且 mid 又比開頭更大 (0 &rarr; mid 遞增數列), 說明前半的數字太大了, 應該往後面找

如果 mid 較小, 結尾也較 target 小 且 mid 又比結尾更小 (mid &rarr; -1 遞增數列), 說明後半的數字太小了, 應該往前面找

Time: O(log N)

````python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        if not nums: return -1

        lo, hi = 0, len(nums) - 1
        
        while lo <= hi:
            mid = (lo + hi) // 2
            
            if target == nums[mid]:
                return mid
            
            if nums[mid] > target:
                # front half is bigger
                if nums[0] > target and nums[mid] >= nums[0]:
                    lo = mid + 1
                else:
                    hi = mid - 1

            if nums[mid] < target:
                # back half is smaller
                if nums[-1] < target and nums[mid] <= nums[-1]:
                    hi = mid - 1
                else:
                    lo = mid + 1
        return -1
````

[300. Longest Increasing Subsequence](https://leetcode.com/problems/longest-increasing-subsequence/)

矩陣 ary 儲存連續遞增的數列

遍歷 nums

當有數字比 ary\[-1] 還要大, 放進 ary 後頭

有數字比 ayr\[0] 還要小, 將 ary\[0] 改成該數字

其他狀況做 binary search 找應該取代的座標

最後回傳 ary 的長度就是最大的連續遞增子數列長度

Time: O(N log N)

````python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        if len(nums) == 0: return 0
        ary = [nums[0]]

        for i in range(1, len(nums)):
            if nums[i] > ary[-1]:
                ary.append(nums[i])
            if nums[i] < ary[0]:
                ary[0] = nums[i]
            else:
                idx = self.insert(ary, nums[i])
                ary[idx] = nums[i]
        return len(ary)

    def insert(self, ary, target):
        lo, hi = 0, len(ary) - 1
        while lo < hi:
            mid = (lo + hi) // 2
            if ary[mid] < target:
                lo = mid + 1
            else:
                hi = mid
        if ary[lo] >= target:
            return lo
        return mid
````

[Minimum Size Subarray Sum](https://leetcode.com/problems/minimum-size-subarray-sum/)

找到最短的連續子陣列, 總和大於 s

### Bianry Search

先建立 cums 紀錄累積總合

找到超越 s 的數字, 定為 right

在左邊用 binary search 找最接近的 left 可以讓 cums\[right] - cums\[left] >- s

Time: O(n log n)

````python
# binary search
class Solution:
    def minSubArrayLen(self, s, nums) -> int:
        n = len(nums)
        if n < 1:
            return 0
        res = float('inf')
        cums = [nums[0]]
        for i in range(1,n):
            cums.append(cums[-1] + nums[i])
        left = 0
        for right, cum in enumerate(cums):
            if cum >= s:
                left = self.search(left, right, cums, s)
                res = min(res, right - left + 1)
        return res if res < float('inf') else 0

    def search(self, lo, hi, cums, target):
        cur = cums[hi]
        while lo < hi:
            mid = (lo+hi) // 2
            # there might be closer one in the right
            if cur - cums[mid] >= target:
                lo = mid + 1
            # not enough, the answer must in the left
            else:
                hi = mid
        
        return lo
````

### 2 pointers

cums: 當前子陣列的總和

遍歷陣列座標為 right, 直到 cums 超過 s, 更新 res, 接著將 nums\[left] 減掉, left 前進

Time: O(n)

````python
# 2 pointers
class Solution:
    def minSubArrayLen(self, s: int, nums: List[int]) -> int:
        n = len(nums)
        if n < 1:
            return 0
        cums = nums[0]
        if cums>=s:
            return 1
        res = float('inf')
        left = 0
        for right, num in enumerate(nums[1:], 1):
            cums += num
            while cums >= s:
                res = min(res, right - left + 1)
                cums -= nums[left]
                left += 1
        return res if res<float('inf') else 0
````

[Search a 2D Matrix](https://leetcode.com/problems/search-a-2d-matrix/)

將 2d array 當作一個長度為 n * m - 1 的 array

row = idx // n, col = idx % n

Time: O(log (mn))

Space: O(1)

````python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        m = len(matrix)
        if m == 0:
            return False
        
        n = len(matrix[0])
        if n == 0:
            return False
        
        lo, hi = 0, n*m-1
        while lo <= hi:
            mid = (lo+hi) // 2
            row = mid //n
            col = mid % n
            if matrix[row][col] < target:
                lo = mid + 1
            elif matrix[row][col] > target:
                hi = mid - 1
            else:
                return True
        return False
````

````python
class Solution:
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        # start from the most bottom-left element
        if len(matrix) == 0 or len(matrix[0]) == 0:
            return False
        
        row = len(matrix) - 1
        col = 0
        
        while row >= 0 and col < len(matrix[0]):
            # print(matrix[row][col], row, col)
            if target > matrix[row][col]:
                col += 1
            elif target < matrix[row][col]:
                row -= 1
            else:
                return True

        return False
        
````

[Pow(x, n)](https://leetcode.com/problems/powx-n/)

res = x * x * ... * x

單純相乘 Time: O(n)



recursive:

每次遞迴處理一半 binary search

base case: n == 0, 結果為 1

特殊狀況: n < 0, 結果是 1.0 / x^(-n)

n 為偶數: half * half

n 為奇數: x \* half \* half

Time: O(log n)

````python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        
        if n == 0: return 1
        if n < 0:
            return 1.0 / self.myPow(x, -1 * n)

        b = n // 2
        half = self.myPow(x, b)
        
        if n % 2 == 0:
            res = half * half
        else:
            res = x * half * half
        
        return res
````

先將所有 row 的開頭推進 heap 中

迴圈 k 次:

將 heap 中最小的元素丟出來, 找到同一列的下一個元素, 推進 heap 中

最後就能得到答案

Time: O(min(k, n) + k log x)

+ heapify: O(min(k, n))
+ k iteration: both heappop and heappush are O(log x)

Space: O(min(k, n))


````python
import heapq
class Solution:
    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:
        
        n = len(matrix)
        
        minHeap = []
        
        # push the first element of each row into the heap
        for r in range(min(k, n)):
            minHeap.append((matrix[r][0], r, 0))
        
        heapq.heapify(minHeap)
        
        for _ in range(k):
            
            value, r, c = heapq.heappop(minHeap)
            
            # still have next value in the same row
            if c + 1 < n:
                heapq.heappush(minHeap, (matrix[r][c+1], r, c+1))
            
        return value
````

#### sorted matrix: binary search

binary search on number range

類似 [Search a 2D Matrix](https://leetcode.com/problems/search-a-2d-matrix/)

1. lo = matrix\[0]\[0], hi = matrix\[n-1]\[n-1]
2. mid = (hi - lo) / 2, 不一定存在於 matrix 中
3. 計算有幾個數字小於等於 mid
4. 計算過程中要記得最靠近 mid 的兩個數字, 略小於或等於 mid 的數字 (mid-), 及略大於 mid 的數字 (mid+)
5. 如果剛好等於 k, 那 mid 就是結果
6. 如果小於 k, hi = mid-; 如果大於 k, lo = mid+

[Find Minimum in Rotated Sorted Array](https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/)

最小值一定是反轉點, 兩側都比它大

nums[mid] 總是與 nums[hi] 比較

mid 較小: 說明右側是排序好的 array, 要去左邊找 min

mid 較大: 說明左側是排序好的 array, 要去右邊找 min

Time: O(log n)

Space: O(1)

````python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        if not nums:
            return -1
        
        lo, hi = 0, len(nums)-1
        
        if (nums[lo] < nums[hi]):
            # sorted array
            return nums[lo]
        
        while lo < hi:
            mid = (hi+lo) // 2
            if nums[mid] < nums[hi]:
                hi = mid
            else:
                lo = mid + 1
        
        return nums[lo]
````

[29. Divide Two Integers]https://leetcode.com/problems/divide-two-integers/)


兩個表格

powersOfTwo: 儲存 2 的次方, 之後從這邊調用商數

doubles: 儲存 divisor * 2的次方

從後面遍歷 doubles, 遇到比 dividend 小的元素

dividend -= double\[i]

quotient += powersOfTwo

Time: O(log n)

Space: O(log n)

````python
class Solution:
    def divide(self, dividend: int, divisor: int) -> int:
        MAX_INT = 2147483647
        MIN_INT = -2147483648
        HALF_MIN_INT = -1073741824
        
        
        if dividend == MIN_INT and divisor == -1:
            return MAX_INT

        negatives = 2
        if dividend > 0:
            negatives -= 1
            dividend = -dividend
        if divisor > 0:
            negatives -= 1
            divisor = -divisor

        doubles = []
        powersOfTwo = []

        powerOfTwo = 1
        while divisor >= dividend:
            doubles.append(divisor)
            powersOfTwo.append(powerOfTwo)
            
            if divisor < HALF_MIN_INT:
                break
            divisor += divisor
            powerOfTwo += powerOfTwo

        quotient = 0
        for i in reversed(range(len(doubles))):
            if doubles[i] >= dividend:
                quotient += powersOfTwo[i]
                dividend -= doubles[i]

        return quotient if negatives != 1 else -quotient
````
