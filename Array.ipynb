# [Minimum Size Subarray Sum](https://leetcode.com/problems/minimum-size-subarray-sum/)

找到最短的連續元素和 大於等於目標

## O(n)

sliding window

找到可以取得目標值得頭跟尾

尾往前

大於等於目標值: 頭往前

## O(nlogn)



## O(N^2) TLE

遍歷矩陣

從某元素出發 遍歷剩下的矩陣 直到 sum >= s

記錄下當時的長度 如果走到底了 就回傳 0

也可以搞成迭代 nums[index: ]

```
def minSubArrayLen_TLE(s, nums):
    res = float('inf')
    for i in range(len(nums)):
        n_sum = 0
        for j in range(i ,len(nums)):
            n_sum += nums[j]
            if n_sum >= s:
                tmp = (j - i + 1)
                if tmp < res: res = tmp
                break
    return res if res != float('inf') else 0
```

```
# O(n), 96 ms
def minSubArrayLen(s, nums):
    res = len(nums) + 1 ## 不會超過矩陣長度
    start, total = 0, 0
    for end, num in enumerate(nums):
        total += num
        while total >= s:
            res = min(res, end - start + 1)
            total -= nums[start]
            start += 1
    return res if res <= len(nums) else 0
```

```
## O(nlogn)
def minSubArrayLen(target, nums):
    result = len(nums) + 1
    for idx, n in enumerate(nums[1:], 1):
        nums[idx] = nums[idx - 1] + n
    left = 0
    for right, n in enumerate(nums):
        if n >= target:
            left = find_left(left, right, nums, target, n)
            result = min(result, right - left + 1)
    return result if result <= len(nums) else 0

def find_left(left, right, nums, target, n):
    while left < right:
        mid = (left + right) // 2 ## log n
        if n - nums[mid] >= target:
            left = mid + 1
        else:
            right = mid
    return left
```

```
## O(n), 72 ms
def minSubArrayLen(self, target: int, nums: List[int]) -> int:
    if len(nums) == 0:
        return 0
    if target == 0:
        return 1
    start, end, maxLen = 0, 0, 0
    sum = 0
    for i in range(0, len(nums)):
        sum += nums[i]
        if sum >= target:
            while sum >= target:
                sum -= nums[start]
                start += 1
            if maxLen == 0 or maxLen > (i - start + 2):
                maxLen = i - start + 2
    return maxLen
```

# [Rotate Array](https://leetcode.com/explore/learn/card/array-and-string/204/conclusion/1182/)

越多解法越好

1.   遍歷矩陣 把數字塞到新的矩陣 (index+k) % len 的格子去

2.   切成兩塊在合成: 超過長度的一塊 還在裡面的一塊 => inplace

3.   轉成 deque 再轉回來 => inplace

4.   黏上一個一樣的 array 在後面 取 [k-1:len+k]


```
def rotate(nums, k):
    k = k % len(nums)
    nums[:] = nums[-k:] + nums[:-k]
```

```
rotate([1,2,3,4,5,6,7], 3)
```

# [Pascal's Triangle II](https://leetcode.com/explore/learn/card/array-and-string/204/conclusion/1171/)

只要抓對應行數就好

只用 O(k) 的空間

倒著計算就不用擔心 覆蓋到需要的數字

```
def getRow(rowIndex):
    res = [1] * (rowIndex + 1)
    for i in range(1, rowIndex + 1):
        for j in range(i-1, 0, -1):
            res[j] += res[j-1]
    return res
```

```
## less space
def getRow(rowIndex):
    row = [1]
    for i in range(rowIndex):
        row = [x + y for x, y in zip([0]+row, row+[0])]
    return row
```

```
## faster
def getRow(rowIndex):
    if rowIndex == 0: return [1]
    if rowIndex == 1: return [1, 1]
    else:
        sub = getRow(rowIndex - 1)
        ans = [1]
        for index in range(0, len(sub) - 1):
            ans.append(sub[index] + sub[index + 1])
        ans.append(1)
        return ans
```

# [Reverse Words in a String](https://leetcode.com/explore/learn/card/array-and-string/204/conclusion/1164/)

還要去掉多餘的空格

```
def reverseWords(s):
    words = [word.strip() for word in s.split(" ")]
    head = 0
    tail = len(words) - 1
    while head < tail:
        words[head], words[tail] = words[tail], words[head]
        head += 1
        tail -= 1
    return ' '.join(word for word in words if word != '') 
```

```
reverseWords("  hello world!  ")
```

# [Reverse Words in a String III](https://leetcode.com/explore/learn/card/array-and-string/204/conclusion/1165/)

個別單字內的順序都要改變

字跟字之間順序不變

Input: "Let's take LeetCode contest"

Output: "s'teL ekat edoCteeL tsetnoc"

```
def reverseWords(s):
    words = [word.strip() for word in s.split(" ") if word != '']
    res = []
    for word in words:
        word = list(word)
        head, tail = 0, len(word) - 1
        while head < tail:
            word[head], word[tail] = word[tail], word[head]
            head += 1
            tail -= 1
        word = ''.join(word)
        res.append(word)
    return ' '.join(res)
```

```
def reverseWords(s):
    return ' '.join(word[::-1] for word in s.split(" "))
```

```
reverseWords("Let's take LeetCode contest")
```

# [Remove Duplicates from Sorted Array](https://leetcode.com/explore/learn/card/array-and-string/204/conclusion/1173/)

類似 Remove Element 必須 inplace

前ptr跑比較快 把沒有重複的東西丟到後ptr中

需要一個東西紀錄目前的數字(back指著的東西)

```
def removeDuplicates(nums):
    front = 0
    back = 0
    for front in range(1, len(nums)):
        if nums[front] != nums[back]:
            back += 1
            nums[back] = nums[front]
            print(nums)
    return back + 1
```

```
removeDuplicates([0,0,1,1,1,2,2,3,3,4])
```

# [Move Zeroes](https://leetcode.com/explore/learn/card/array-and-string/204/conclusion/1174/)

front 往前

碰到不是零的: 放進 back 中

最後把back+1到最後一個都變成零


```
def moveZeroes(nums):
    front = 0
    back = 0
    for front in range(len(nums)):
        if nums[front] != 0:
            nums[back] = nums[front]
            back += 1
            print(nums)
    for idx in range(back, len(nums)):
        nums[idx] = 0
    print(nums)
    return back
```
